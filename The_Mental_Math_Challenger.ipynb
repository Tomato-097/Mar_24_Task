```python
import random

score = 0
is_correct = True

while is_correct:
    num1 = random.randint(1, 12)
    num2 = random.randint(1, 12)
    actual_result = num1 * num2
    
    # Type casting the input to an int for comparison
    user_answer = int(input(f"What is {num1} x {num2}? "))
    
    if user_answer == actual_result:
        score += 1
        print(f"Correct! Score: {score}")
    else:
        print(f"Wrong! The answer was {actual_result}.")
        is_correct = False

print(f"Game Over. Final Streak: {score}")
```
