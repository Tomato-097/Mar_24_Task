```python
# The Scenario: You act as a university data analyst. 
#     The legacy admissions portal crashed during a system migration, 
#     and the backup applicant data is a mess. You must clean the 
#     "Applicant Records" dataset.

import pandas as pd

df = pd.read_csv('admissions_messy.csv')

print(df.head())
print(df.info())
df.shape[0]
```

```python
# Handling Missing Data: The dataset is riddled with NaNs. You must use 
#     .dropna() to remove rows completely missing an Application ID.

df_dropna = df.dropna(subset = ['Application ID'])
print("Remove rows completely missing an Application ID")
print(df_dropna.head(10))
df_dropna.shape[0]
```

```python
# You will use .fillna() to replace missing Intended Major and High 
# School State values with "Unknown" that

df_fillna = df_dropna.fillna({'Intended Major': 'Unknown', 'High School State': 'Unknown'})
print("\nReplace Nan value with Unknown")
print(df_fillna.head(10))
df_fillna.shape[0]
```

```python
# Handling Duplicates: A server glitch caused applicants to accidentally 
#         submit identical applications twice. You must use .duplicated() 
#         to find identical application rows 

duplicates = df_fillna.duplicated(subset=['Application ID'], keep=False)

print("\nIdentical Transactions found in 'Application ID':")
print(duplicates.head(10))
```

```python
 # and .drop_duplicates(keep='first') to remove the errors 
 # and keep only the original submission.

cleaned_df =df_fillna.drop_duplicates(subset=['Application ID'], keep='first')

print("\nDataset after removing duplicate Application ID:")
print(cleaned_df.head(10))
cleaned_df.shape[0]
```

```python
# Data Type Conversions: 
#     Application Fee is imported as a string (e.g., "$150.50"). 
#     You must remove the symbols and use pd.to_numeric() to turn 
#     it into a float.

cleaned_df['Application Fee'] = cleaned_df['Application Fee'].str.replace('$', '', regex=False).str.replace(',', '', regex=False)

cleaned_df['Application Fee'] = pd.to_numeric(cleaned_df['Application Fee'])

print(cleaned_df[['Application ID', 'Application Fee']].head(10))
print(f"\nNew Data Type: {cleaned_df['Application Fee'].dtype}")
```

```python
# Replace “pending” and “unknown” text entries in the dates with 
#         the previous valid datetime (forward filling).
# Convert Submission Date from messy text into a proper datetime 
#         object using pd.to_datetime().

targets = ['pending', 'unknown']

cleaned_df['Submission Date'] = cleaned_df['Submission Date'].replace(targets, None)

cleaned_df['Submission Date'] = pd.to_datetime(cleaned_df['Submission Date'], format='mixed', errors='coerce')

cleaned_df['Submission Date'] = cleaned_df['Submission Date'].ffill().bfill()

print(cleaned_df[['Submission Date']].head(10))
```

```python
# Convert the Submission Date format from various mixed formats into a 
# uniform "M/D/Y" format.

cleaned_df['Submission Date'] = cleaned_df['Submission Date'].dt.strftime('%m/%d/%Y')

print(cleaned_df[['Submission Date']].head(10))
```
