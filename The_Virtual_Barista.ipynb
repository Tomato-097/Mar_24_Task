```python
menu = {"Coffee": 3.0, "Tea": 2.5, "Latte": 4.5}
total = 0.0

print("--- MENU ---")
print("Coffee: $3.0 | Tea: $2.5 | Latte: $4.5")
print("(Type 'checkout' to finish)")

while True:
    order = input("\nWhat would you like? ").lower()

    if order == "checkout":
        break

    if order == "coffee":
        total += menu["Coffee"]
    elif order == "tea":
        total += menu["Tea"]
    elif order == "latte":
        total += menu["Latte"]
    else:
        print("Sorry, that's not on the menu.")

print(f"Final total is: ${total:.2f}")
```
