**Dataset:** Synthetic data (manually typed lists).  
**Goal:** Reinforce the Matplotlib Object-Oriented canvas structure and introduce basic Seaborn styling to track how two different medications reduce pain over time.

### 1. Importing Libraries & Theming:
- Start by importing matplotlib.pyplot as plt and import seaborn as sns.  
- New Skill: Before creating your plot, use Seaborn to instantly upgrade the visuals by setting a theme. Call sns.set_theme(style="darkgrid").

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

### 2. The Anatomy / Basic Plotting:
- Create lists to track a 6-hour window.
- hours = [1, 2, 3, 4, 5, 6]
- Create two lists for the medications, representing reported pain levels on a scale of 0 to 10 (e.g., med_A_pain = [8, 6, 5, 3, 2, 2]).

```python
hours = [1, 2, 3, 4, 5, 6]

med_A_pain = [8, 6, 5, 3, 2, 2]
med_B_pain = [9, 8, 7, 5, 4, 4]
```

Set up your canvas using the Object-Oriented method: fig, ax = plt.subplots().  
Plot the data using ax.plot(hours, med_A_pain) and ax.plot(hours, med_B_pain).  
New Skill: Inside your ax.plot() calls, try adding markers and specific colors to differentiate the lines further (e.g., ax.plot(..., color='blue', marker='o', label='Medication A')).


```python
fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(hours, med_A_pain, color='blue', marker='o', label='Medication A')
ax.plot(hours, med_B_pain, color='green', marker='s', label='Medication B')

plt.show()
```

### 3. Basic Customization:
Instruct the canvas to display your metadata:
- ax.set_title("Pain Level Over Time: Med A vs. Med B")
- ax.set_xlabel("Hours Post-Dose")
- ax.set_ylabel("Reported Pain Level (0-10)")  
Call ax.legend() so the viewer knows which line represents which medication.
Finally, render the plot cleanly with plt.show().

```python
ax.set_title("Pain Level Over Time: Med A vs. Med B")
ax.set_xlabel("Hours Post-Dose")
ax.set_ylabel("Reported Pain Level (0-10)")
ax.legend()

plt.show()
```
