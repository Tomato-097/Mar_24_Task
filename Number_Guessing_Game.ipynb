```python
import random

target = random.randint(1, 100)
guess = 0 

while guess != target:
    guess = int(input("Guess (1-100): "))

    if guess > target:
        print("Too high!")
    elif guess < target:
        print("Too low!")
    else:
        print(f"You got it! The number was {target}.")

print("Game Over.")
```
